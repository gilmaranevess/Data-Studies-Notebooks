```python
import seaborn as sns
%matplotlib inline 
#to see the graphics inside jupyter notebook
#seaborn comes with some data sets 
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=40)

#kde to remove the line in the graphic
#bins to show more distribution
```

```python
#sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
#regression

sns.jointplot(x='total_bill',y='tip',data=tips, kind='kde')
#where they match up the most
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
#hue for categorical columns / not numerical
#pass your dataframe
#relationship across an intire data frame for the numerical columns
```

```python
sns.rugplot(tips['total_bill'])

#rugplot - distraws a dashmark for every point of any variable distribution
```

```python
sns.distplot(tips['total_bill'])

# kernel density estimation plot - kde
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips,
           estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
#not for very large datasets
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```
