```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL
plt.plot(x,y)
#plt.show() use this if you are not using jupyter notebook
```

```python
plt.plot(x,y,'r-')
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
#MultiPlots in the same campus
plt.subplot(1,2,1)#number of rows, columns, plot number u are refering to
plt.plot(x,y,'r')

plt.subplot(1,2,2) #1 row by 2 columns, second plot
plt.plot(y,x,'b')
```

```python
#API METHODS
#OO object oriented method to have more control over the graphic

fig = plt.figure() #using OO

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left of the axes, bottom, width and high 
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

#all the values of the axes have to be less than 1
#left, bottom, width and high
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.15,0.52,0.45,0.25])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
#to build 1 row with 2 columns
fig,axes = plt.subplots(nrows=1, ncols=2)
#plt.tight_layout()

for current_ax in axes:
    current_ax.plot(x,y)

#axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
#FIGURE SIZE AND DPI
```

```python
fig = plt.figure(figsize=(8,2), dpi=100)
#8 inches by 2 inches

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

#axes.plot(x,y)
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label= 'X Cubed')

ax.legend(loc='best') #choose the best location

ax.legend(loc=(0.1,0.1)) 
```

```python
#PLOT APPEARANCE
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#FF8C00', linewidth=5, alpha=0.5) 
#RGB HEX CODES
#linewidth or lw - line thickness
#alpha - line transparency
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=5, linestyle=':') 
#linestyle or ls = steps, - 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=1, ls='-', marker='o', 
        markersize='5', markerfacecolor='yellow', 
        markeredgewidth=3, markeredgecolor='green') 
```
